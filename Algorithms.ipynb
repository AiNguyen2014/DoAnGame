```python
import matplotlib.pyplot as plt
import numpy as np

algorithms = ['A*', 'BFS', 'Q-Learning', 'Min-Conflict', 'Local Beam Search']
levels = ['Level 1', 'Level 4', 'Level 6']
times = [[18, 22, 95, 28, 25], [32, 38, 115, 42, 40], [48, 52, 145, 58, 55]]  # Thi gian x l (ms)
success_rates = [[96, 94, 82, 90, 92], [88, 85, 68, 82, 84], [78, 75, 58, 72, 74]]  # T l thnh cng (%)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Biu  thi gian x l
for i, level in enumerate(levels):
    ax1.bar(np.arange(len(algorithms)) + i*0.15, times[i], width=0.15, label=level)
ax1.set_xticks(np.arange(len(algorithms)))
ax1.set_xticklabels(algorithms, rotation=45)
ax1.set_ylabel('Thi gian x l (ms)')
ax1.set_title('Thi gian x l trung bnh theo cp ')
ax1.legend()

# Biu  t l thnh cng
for i, level in enumerate(levels):
    ax2.plot(algorithms, success_rates[i], marker='o', label=level)
ax2.set_ylabel('T l thnh cng (%)')
ax2.set_title('T l thnh cng theo cp ')
ax2.legend()

plt.tight_layout()
plt.savefig('performance_comparison.png')
plt.show()
```
