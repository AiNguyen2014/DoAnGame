```python
import matplotlib.pyplot as plt
import numpy as np

algorithms = ['A*', 'BFS', 'Q-Learning', 'Min-Conflict', 'Local Beam Search']
levels = ['Level 1', 'Level 4', 'Level 6']
times = [[18, 22, 95, 28, 25], [32, 38, 115, 42, 40], [48, 52, 145, 58, 55]]  # Thời gian xử lý (ms)
success_rates = [[96, 94, 82, 90, 92], [88, 85, 68, 82, 84], [78, 75, 58, 72, 74]]  # Tỷ lệ thành công (%)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Biểu đồ thời gian xử lý
for i, level in enumerate(levels):
    ax1.bar(np.arange(len(algorithms)) + i*0.15, times[i], width=0.15, label=level)
ax1.set_xticks(np.arange(len(algorithms)))
ax1.set_xticklabels(algorithms, rotation=45)
ax1.set_ylabel('Thời gian xử lý (ms)')
ax1.set_title('Thời gian xử lý trung bình theo cấp độ')
ax1.legend()

# Biểu đồ tỷ lệ thành công
for i, level in enumerate(levels):
    ax2.plot(algorithms, success_rates[i], marker='o', label=level)
ax2.set_ylabel('Tỷ lệ thành công (%)')
ax2.set_title('Tỷ lệ thành công theo cấp độ')
ax2.legend()

plt.tight_layout()
plt.savefig('performance_comparison.png')
plt.show()
```

|---------------------|--------------|--------------|--------------|-------------------------------|-------------------------------|-------------------------------|
